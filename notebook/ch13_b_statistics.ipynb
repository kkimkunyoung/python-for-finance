## 13.3 베이즈 통계학

### 13.3.1 베이즈 공식

금융 분야에서는 베이즈 공식을 해석하는 데 **통시적 방법**(diachronic interpretation)을 사용.

: 시간이 지남에 따라 시계열의 평균 수익률처럼 관심을 가진 변수나 모수에 대한 새로운 정보를 학습한다는 관점.

베이즈 공식

$p(H | D) = \dfrac{p(H) \cdot p(D | H)}{p(D)}$

- $H$는 사건 혹은 가정, $D$는 현실에서 얻은 데이터

- $p(H)$: 사전확률 prior probability

- $p(D)$: 해당 데이터를 얻을 확률. 정규화 상수 normalizing constant.

- $p(D|H)$: 가능도 likelihood. 가정 $H$하에서 데이터를 얻을 확률

- $p(H|D)$: 사후확률 posterior probability. 데이터를 얻은 뒤의 확률.

$사전 \, 확률 + 증거 \, 관찰 \rightarrow "사후 \, 확률"$

### 13.3.2 베이즈 회귀

```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
np.random.seed(1000)
%matplotlib inline
```

```python
x = np.linspace(0, 10, 500)
y = 4 + 2 * x + np.random.standard_normal(len(x)) * 2
```

```python
reg = np.polyfit(x, y, 1)
```

```python
reg
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, c=y, marker='v', cmap='coolwarm')
plt.plot(x, reg[1] + reg[0] * x, lw=2.0)
plt.colorbar()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
```

최소 자승 회귀법의 결과는 회귀선의 두 모수(기울기와 절편)에 대한 고정된 값.

데이터에 더해진 잡음 때문에 원래의 모숫값 2와 4를 완벽하게 복원하지는 못함.
