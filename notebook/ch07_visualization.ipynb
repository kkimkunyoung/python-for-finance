```python
def func(x):
    return 0.5 * np.exp(x) + 1
    
a, b = 0.5, 1.5

x = np.linspace(0,2)
y = func(x)

Ix = np.linspace(a, b)
Iy = func(Ix)

verts = [(a,0)] + list(zip(Ix, Iy)) + [(b,0)]
```

```python
from matplotlib.patches import Polygon

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(bottom=0)

poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)

plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x) \mathrm{d}x$',
         horizontalalignment='center', fontsize=20)

plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks((func(a), func(b)))
ax.set_yticklabels(('$f(a)$', '$f(b)$'))

plt.show()
```

## 7.2 정적 3차원 플롯

```python
strike = np.linspace(50, 150, 24)
```

```python
ttm = np.linspace(0.5, 2.5, 24)
```

```python
strike, ttm = np.meshgrid(strike, ttm)
```

```python
strike[:2].round(1)
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm
```

```python
iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(projection='3d')

surf = ax.plot_surface(strike, ttm, iv,
                      rstride=2, cstride=2,
                      cmap=plt.cm.coolwarm, linewidth=0.5,
                      antialiased=True)

ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,6))

ax = fig.add_subplot(projection='3d')

ax.view_init(30, 60)

ax.scatter(strike, ttm, iv,
           zdir='z', s=25, c='b', marker='^')

ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')

plt.show()
```

## 7.3 상호작용형 2차원 플롯

### 7.3.1 기본 플롯

```python
import pandas as pd

a = np.random.standard_normal((250,5)).cumsum(axis=0)

index = pd.date_range('2019-1-1', 
                      freq='B', 
                      periods=len(a))

df = pd.DataFrame(100 + 5 * a,
                  columns=list('abcde'),
                  index=index)

df.head()
```

```python
import plotly.express as px

fig = px.line(df)

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df['a'],
        mode='markers',
        name='a',
        marker=dict(
            symbol='circle',
            size=3.5,
            color='blue'
        )
    )
)

fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df['b'],
        mode='lines+markers',
        name='b',
        line=dict(
            color='magenta'
        ),
        marker=dict(
            symbol='diamond',
            size=3.5,
            color='magenta'
        )
    )
)

fig.update_layout(
    title='A Time Series Plot',
    xaxis_title='date',
    yaxis_title='value'
)

fig.show()
```

```python
from plotly.subplots import make_subplots

rows=3
cols=2

fig = make_subplots(
    rows=rows,
    cols=cols
)

for i, column in enumerate(df.columns):
    row = i // cols + 1
    col = i % cols + 1

    fig.add_trace(
        go.Histogram(
            x=df[column],
            nbinsx=15,
            name=column
        ),
        row=row,
        col=col
    )

fig.update_layout(
    height=250 * rows,
    bargap=0.1
)

fig.show()
```

### 7.3.2 금융 플롯

```python
raw = pd.read_csv('../source/fxcm_eur_usd_eod_data.csv',
                  index_col=0, parse_dates=True)
```

```python
raw.info()
```

```python
quotes = raw[['AskOpen', 'AskHigh', 'AskLow', 'AskClose']]
quotes = quotes.iloc[-60:]
quotes.tail()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Candlestick(
        x=quotes.index,
        open=quotes['AskOpen'],
        high=quotes['AskHigh'],
        low=quotes['AskLow'],
        close=quotes['AskClose'],
        name='EUR/USD'
    )
)

fig.update_layout(
    title='EUR/USD Exchange Rate',
    height = 500,
    xaxis_rangeslider_visible=False
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Candlestick(
        x=quotes.index,
        open=quotes['AskOpen'],
        high=quotes['AskHigh'],
        low=quotes['AskLow'],
        close=quotes['AskClose'],
        name='EUR/USD'
    )
)

WINDOW=15
STD=2

close = quotes['AskClose']

middle = close.rolling(window=WINDOW).mean()
std = close.rolling(window=WINDOW).std()

upper = middle + STD * std
lower = middle - STD * std

fig.add_trace(go.Scatter(x=quotes.index, y=upper, mode='lines', name='Upper'))
fig.add_trace(go.Scatter(x=quotes.index, y=middle, mode='lines', name='BOLL(AskClose, 15)'))
fig.add_trace(go.Scatter(x=quotes.index, y=lower, mode='lines', name='Lower'))

fig.update_layout(
    title='EUR/USD Exchange Rate',
    height=500,
    xaxis_rangeslider_visible=False
)

fig.show()
```

```python
fig = make_subplots(
    rows=2,
    cols=1,
    shared_xaxes=True,
    row_heights=[0.8, 0.2],
    vertical_spacing=0.05
)

# 1. Candlestick
fig.add_trace(
    go.Candlestick(
        x=quotes.index,
        open=quotes['AskOpen'],
        high=quotes['AskHigh'],
        low=quotes['AskLow'],
        close=quotes['AskClose'],
        name='EUR/USD'
    ),
    row=1,
    col=1
)

# 2. Bollinger Bands
close = quotes['AskClose']

middle = close.rolling(window=15).mean()
std = close.rolling(window=15).std()

upper = middle + 2 * std
lower = middle - 2 * std

fig.add_trace(go.Scatter(x=quotes.index, y=upper, mode='lines', name='Upper'), row=1, col=1)
fig.add_trace(go.Scatter(x=quotes.index, y=middle, mode='lines', name='BOLL(AskClose,15)'), row=1, col=1)
fig.add_trace(go.Scatter(x=quotes.index, y=lower, mode='lines', name='Lower'), row=1, col=1)

# 3. RSI
delta = close.diff()

gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)

avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()

rs = avg_gain / avg_loss
rsi = 100 - (100 / (1 + rs))

fig.add_trace(
    go.Scatter(
        x=quotes.index,
        y=rsi,
        mode='lines',
        name='RSI(14)'
    ),
    row=2,
    col=1
)

fig.update_layout(
    title='EUR/USD Exchange Rate',
    height=700,
    xaxis_rangeslider_visible=False
)

fig.show()
```
